There are also two different ways to incorporate the HTML, such as:
- app = FastHTML() 
    - This allows for specifications within routing and retrieving the website 
    - ex: "@app.get('/')", "@app.{get, post, put, patch, delete, route}('/')", "@app.route('/', methods=['GET', 'POST'])"
- app, rt = fast_app() 
    - Allows for shortened route names and allows methods
    - ex: "@rt('/', methods=['post'])"


```python
app = FastHTML()

@app.get('/')
def Home():
    return P("Hello World!")

serve() # Used when setting up server | recommended using in the empty py file given, it will not work in jupyter
```

```python
app, rt = fast_app()

@rt('/')
def Home():
    return P("Hello World!")

serve()
```
