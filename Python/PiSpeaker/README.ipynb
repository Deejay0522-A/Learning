This code should be able to power a component connected to a pin and a ground.

```python
import RPi.GPIO as GPIO #type: ignore
#Remove the type command to make sure you have RPi installed
import time

# Copy the setup and output for as many pins you have set.

GPIO.setmode(GPIO.BCM) # Or board
GPIO.setup(26, GPIO.OUT) # Out for an output, In for input usage
GPIO.output(26, GPIO.HIGH) # pin, output mode (High[ON], Low[OFF])

time.sleep(2)

GPIO.output(26, GPIO.LOW)

GPIO.cleanup() # Resets the pins
```
